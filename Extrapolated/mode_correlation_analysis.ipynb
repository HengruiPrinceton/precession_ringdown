# wigner-D Matrix

```python
from sphere_hz import *
def lm_index(l,m):
    return l**2+l+m-4
```

```python
# Construct Grid
Ntheta = 20 #80
Nphi = 2*Ntheta
nmax = 10
t,p,T,P,ws = grid(Ntheta,Nphi)

# Construct Basis function
D,D_t,D_p,D_tt,D_pp,D_tp = compute_bases(nmax,T,P)

# Convert to cartesian basis
xyz = to_xyz(T,P)


# Construct spin-weighted basis
lmax=8

m2D = []
for i in range(4,(lmax+1)**2):
    m2D.append(lowerspin(D[i],T,ws,D,D_t,D_p,D_tt,D_pp,D_tp))
#for i in range((lmax+1)**2):
m2D = np.array(m2D)

normalization = []
for i in range(m2D.shape[0]):
    normalization.append(inner_product(m2D[i],m2D[i],ws))
normalization = np.array(normalization)

m2D /=np.sqrt(normalization[:,np.newaxis,np.newaxis])#/np.sqrt(2)
```

```python
thetas = np.linspace(0.01,np.pi/2,200)
r22_21 = []
r22_20 = []

r33_32 = []
r33_31 = []
r33_30 = []

r44_43 = []
r44_42 = []
r44_41 = []
r44_40 = []
for theta_offset in thetas:
    # Then rotate along the ROTATED y axis to align theta
    xyz_rot = rotation_y(xyz,-theta_offset)
    T_rot,P_rot = to_thetaphi(xyz_rot)

    # scalar spherical harmonics in rotated frame
    D2,D2_t,D2_p,D2_tt,D2_pp,D2_tp = compute_bases(nmax,T_rot,P_rot)

    # Construct spin-weighted basis
    lmax=8

    m2D_rot = []
    for j in range(4,(lmax+1)**2):
        m2D_rot.append(lowerspin(D2[j],T,ws,D,D_t,D_p,D_tt,D_pp,D_tp))
        #normalizing
        m2D_rot[-1]/=np.sqrt(inner_product(m2D_rot[-1],m2D_rot[-1],ws))
    m2D_rot = np.array(m2D_rot)

    # Jacobian for bases tramsformation
    J = basis_transform(m2D,m2D_rot,ws)
    r22_21.append((J[lm_index(2,2),lm_index(2,1)]+J[lm_index(2,-2),lm_index(2,1)])/
                 (J[lm_index(2,2),lm_index(2,2)]+J[lm_index(2,-2),lm_index(2,2)]))
    r22_20.append((J[lm_index(2,2),lm_index(2,0)]+J[lm_index(2,-2),lm_index(2,0)])/
                 (J[lm_index(2,2),lm_index(2,2)]+J[lm_index(2,-2),lm_index(2,2)]))
    
    #r33_32.append((J[lm_index(3,3),lm_index(3,2)]+J[lm_index(3,-3),lm_index(3,2)])/
    #             (J[lm_index(2,2),lm_index(2,2)]+J[lm_index(2,-2),lm_index(2,2)]))
    #r33_31.append((J[lm_index(3,3),lm_index(3,1)]+J[lm_index(3,-3),lm_index(3,1)])/
    #             (J[lm_index(2,2),lm_index(2,2)]+J[lm_index(2,-2),lm_index(2,2)]))
    #r33_30.append((J[lm_index(3,3),lm_index(3,0)]+J[lm_index(3,-3),lm_index(3,0)])/
    #             (J[lm_index(2,2),lm_index(2,2)]+J[lm_index(2,-2),lm_index(2,2)]))
    
    #r44_43.append((J[lm_index(4,4),lm_index(4,3)]+J[lm_index(4,-4),lm_index(4,3)]))
    #r44_42.append((J[lm_index(4,4),lm_index(4,2)]+J[lm_index(4,-4),lm_index(4,2)]))
    #r44_41.append((J[lm_index(4,4),lm_index(4,1)]+J[lm_index(4,-4),lm_index(4,1)]))
    #r44_40.append((J[lm_index(4,4),lm_index(4,0)]+J[lm_index(4,-4),lm_index(4,0)]))
    #print(theta_offset)
r22_21 = np.array(r22_21)
r22_20 = np.array(r22_20)

#r33_32 = np.array(r33_32)
#r33_31 = np.array(r33_31)
#r33_30 = np.array(r33_30)

#r44_43 = np.array(r44_43)
#r44_42 = np.array(r44_42)
#r44_41 = np.array(r44_41)
#r44_40 = np.array(r44_40)
```
